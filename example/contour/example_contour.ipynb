```python
from ternary_diagram import TernaryDiagram
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('example_contour.csv')
df.head()
```

Compounds corresponding to vertices

Specify triangle labels around the half-hour mark.

```python
materials = df.columns[0:3].tolist()
materials
```

```python
fig = plt.figure(facecolor='white')
td = TernaryDiagram(materials=materials)
td.contour(df[materials], z = df['z'], z_min = 0, z_max = 1, bar_label = 'example_contour')
```

```python
fig.savefig('example_contour.png', dpi = 144)
```
