```python
from ternary_diagram import TernaryDiagram
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('example_mono_scatter.csv')
df.head()
```

### Plot

```python
fig = plt.figure(facecolor='white')
td = TernaryDiagram(df.columns)
td.scatter(df, color = 'blue', facecolor = "None", lw = 2)
```

### Tie line

```python
fig = plt.figure(facecolor='w')
td = TernaryDiagram(df.columns)
td.scatter(df.iloc[:10], label = '<10')
td.scatter(df.iloc[10:], label = '>10')
td.plot([[1,0,1], [0,2,3]], label = 'tie line')
td.ax.legend()
fig.tight_layout()
```

```python
fig.savefig('example_mono_scatter.png', dpi = 144)
```
