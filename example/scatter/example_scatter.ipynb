```python
from ternary_diagram import TernaryDiagram
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('example_scatter.csv')
df.head()
```

```python
materials = df.columns[0:3]
```

```python
fig = plt.figure(facecolor='w')
td = TernaryDiagram(materials=materials)
td.scatter(df[materials], z = df['z'], z_min = 0, z_max = 1)
```

```python
td.fig.savefig('example_scatter.png', dpi=144)
```
